```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Product Name'] = le.fit_transform(df['Product Name'])
```

```python
df.head()
```

```python
df.info()
```

```python
corr_matrix = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr_matrix, annot=True, linewidths=2)
```

```python
df['Weather'] = le.fit_transform(df['Weather'])
```

```python
df.head()
```

```python
df['Date'].value_counts()
```

```python
corr_matrix = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr_matrix, annot=True, linewidths=2)
```

```python
df = df.drop(['Date','Weather'],axis = 1)
```

```python
df.head()
```

```python
df['Festival Season'] = le.fit_transform(df['Festival Season'])
```

```python
df.head()
```

```python
df['Festival Season'].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()
df['Stock Availability'] = oe.fit_transform(df[['Stock Availability']])
```

```python
df.head()
```

```python
df['Stock Availability'] = df['Stock Availability'].astype(int)
```

```python
df.head()
```

```python
#to predict price using Regression.

x1 = df.drop(['Price (₹)'],axis = 1)
y1 = df['Price (₹)']
# to predict demand.
x2 = df.drop(['Demand (Units Sold)', 'Google Search Interest','Social Media Mentions','Social Media Mentions','Stock Availability'],axis = 1)
y2 = df['Demand (Units Sold)']
```

```python
from sklearn.model_selection import train_test_split
x1_train,x1_test,y1_train,y1_test = train_test_split(x1,y1,test_size = 0.01,random_state = 42)
x2_train,x2_test,y2_train,y2_test = train_test_split(x2,y2,test_size = 0.01,random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x1_train,y1_train)
```

```python
from sklearn.metrics import r2_score
y1_pred = lr.predict(x1_test)
r2_score(y1_test,y1_pred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y1_test,y1_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(x1_train,y1_train)
```

```python
from sklearn.metrics import r2_score
y1_pred = rfr.predict(x1_test)
r2_score(y1_test,y1_pred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y1_test,y1_pred)
```

```python
from sklearn.model_selection import cross_val_score
cv = cross_val_score(rfr,x1_train,y1_train,cv=10)
```
